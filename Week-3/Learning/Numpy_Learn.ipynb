# Operations in arrays

Arithmetic operations

```python
o1 = np.array([1,2,3,4,5,6])
o2 = np.array([6,2,3,4,5,1])

#Addition
print('Add - ',o1+o2)

#Substraction
print('Subtract - ',o1-o2)

#Multiplication
print('Multiply - ',o1*o2)

#Division
print('Divide - ',o1/o2)

#Mod
print('Mod - ',o1%o2)

#Exponential
print('Exponential - ',o1**o2)
```

Dot Product

```python
np_matrix2 = np.array([[6,4],[2,8]])
print(np.dot(np_matrix, np_matrix2))
```

Trignometry

```python
trig = np.array([0, 15, 30, 45, 60, 90])
trig1 = np.deg2rad(trig)

print("Sin : ", np.sin(trig1))
print("Cos : ", np.cos(trig1))
```

Logarithms

```python
print(np.log(b))
print(np.log10(b))
```

Bitwise operations

```python
bit1 = np.array([1,2,3,4,5], dtype = np.uint8)
bit2 = np.array([9,8,7,6,5], dtype = np.uint8)

print("Bitwise and: ", np.bitwise_and(bit1, bit2))
print("Bitwise or: ", np.bitwise_or(bit1, bit2))
```

# Aggregation Functions : Measuring Central Tendency

```python
arr = np.array([1,2,3,4,5,6])

# Sum of arr
print('Sum - ', np.sum(arr))

# Mean of arr
print('Mean - ',np.mean(arr))

# Median of arr
print('Median - ',np.median(arr))

# Standard Deviation of arr
print('Standard Deviation - ',np.std(arr))

# Variance of arr
print('Variance - ',np.var(arr))

# Maximum value in arr
print('Max - ',np.amax(arr))

# Minimum value in arr
print('Min - ',np.amin(arr))
```

# Broadcasting

Used for arithmetic operations on different sizes of arrays

```python
arr1 = np.array([1,2,3,4,5,6])
arr2 = np.array([[1,2],[3,4]])
print(arr1 + arr2) # Gives error as the shape is not matching
```

Simple broadcasting = scalar multiplication

```python
val = 2
print(arr1*val)
```

If the arrays have different dimension, the shape of the smaller array will be padded onto the bigger array

```python
a = np.array([1,2,3,4,5,6])
b = np.array([[4],[5],[6]])
print(a+b)
```

The size in each dimension must either be equal or one of them must be one.

```python
a = np.array([[1,2,3],[4,5,6]])
b = np.array([4,5,6])
print(a/2)
# Here an array is broadcasted to a matrix
# it happened only because all the arrays had same(3) elements
```

**Limitations of broadcasting**

- Can work only for lower to higher dimension
- Can lead to slow computation if performed incorrectly.

# Matplot Library

```python
import matplotlib.pyplot as plt
```

Line plot

```python
x = np.linspace(0,10, 100)
y = np.sin(x)

plt.plot(x,y)
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Title")
plt.show()
```

Scattered plot

```python
x = np.random.rand(100)
y = np.random.rand(100)

plt.scatter(x,y)
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Title")
plt.show()
```

Histogram

```python
data = np.random.normal(100, 20, 1000) # normal creates a normally distributed data

plt.hist(data, bins=20, color='green')
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Title")
plt.show()
```

Bar Plot

```python
a = ['a', 'b', 'c', 'd', 'e']
b = [10, 20, 30, 40, 50]
# try a different size of a & b

plt.bar(a,b)
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Title")
plt.show()
```

Pie chart

```python
label = ['a', 'b', 'c', 'd', 'e']
value = [10, 20, 30, 40, 50]

plt.pie(value, labels=label, startangle=90, autopct='%1.f%%') #autopct adds %value inside the chart
plt.axis('equal') #both axis will scale equally
plt.title('Pie chart')
plt.show()
```

Box Plot

```python
data = [np.random.normal(50, 10, 100), np.random.normal(80, 20, 100), np.random.normal(40, 15, 100),]

plt.boxplot(data)
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Title")
plt.show()
```

# I/O handling with NumPy

Saving data to a text file

```python
data = np.array([[1,2,3], [7,6,5]])
np.savetxt('datadoc.txt', data, delimiter=',')
# delimiter : seperates the data using the given
```

Saving & Loading data using numpy file

```python
loaded_data = np.loadtxt('datadoc.txt', delimiter=',')
np.save('data.npy', data)
loaded_data = np.load('data.npy')
```

Load data from any text file

```python
data = np.genfromtxt('datadoc.txt', delimiter=',', missing_values=' ', filling_values=0)
data
```

Compressing arrays for efficient storage

```python
np.savez_compressed('data_compressed.npy', data)
```

# Structured Arrays

Defining the your own data type

```python
dt = np.dtype([('name', 'S10'), ('age', 'i4'), ('height', 'f8')])
people = np.array([('John', 32, 1.83), ('Jack', 22, 1.73)], dtype = dt)
people
```

Manuplating data from structed arrays

```python
person = people[0]
person['height']

avg_height = people['height'].mean()
avg_height

older_than_28 = people[people['age'] > 28]
older_than_28
```

```python
# Sorting
sorted_people = np.sort(people, order='age')
sorted_people
```
